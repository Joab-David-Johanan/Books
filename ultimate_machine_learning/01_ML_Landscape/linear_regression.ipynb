```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Download and prepare the data
data_root="https://github.com/ageron/data/raw/main/lifesat/lifesat.csv"
lifesat=pd.read_csv(data_root)
```

Getting the features (X) and labels (y):
- `sklearn` needs features (X) and labels (y) as 2-D arrays, so we use double brackets [[]] to get a DataFrame.
- `lifesat[["column_name"]].values` returns a 2-D array.

```python
X=lifesat[["GDP per capita (USD)"]].values
y=lifesat[["Life satisfaction"]].values
```

```python
lifesat.plot(kind="scatter", x="GDP per capita (USD)", y="Life satisfaction", grid=True)
plt.axis([25_000,60_000,4,9])
plt.show()
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
X_new=[[35_000]]
model.predict(X_new)
```
