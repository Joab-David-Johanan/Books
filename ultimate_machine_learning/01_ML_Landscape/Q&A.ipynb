**1. If I download bunch of data, does that make my computer smart ? What is Machine Learning ?**

- Machine Learning is the field of study that gives computers the ability to learn without being explicitly programmed - coined by Arthur Samuel in 1959.
- A computer program is said to learn from experience `E` with respect to some task `T` and some performance measure `P`, if its performance on task `T` as measured by `P` improves with experience `E`.
- So if you simply download a bunch of wikipedia articles on your computer, then your computer now has lot more data but it has not learned anything.
