***Iterative control statements:*** 
- An iterative control statement is a control statement that is used for repeated execution of a block of statements.
- There are two iterative control statements in python and they are `while` and `for` loop.


While loop is 

```python
# unpacking in for loop
example=[(1,1,1),(2,4,8),(3,9,27),(4,16,64)]
for number, square, cube in example:
    print(number,square,cube)
```

```python
# unpacking in for loop - ignoring one value
example = [(1, 1, 1), (2, 4, 8), (3, 9, 27), (4, 16, 64)]
for number, _, cube in example:
    print(number, cube)
```

```python
# Iterating over dictionaries using for loop
# Remember that dictionaries, sets do not allow direct indexing so for loops work best here
D={'apple':150, 'mango':500, 'banana':350}

# One way to do it
for key in D:
    print(key)
```

```python
# Another way to iterate over the keys of a dictionary
for key in D.keys():
    print(key)

print()

# Way to iterate over the values of a dictionary
for value in D.values():
    print(value)

print()

# Way to iterate over the keys and values of the dictionary using for loop
for item in D.items():
    print(item)

print()

# Whenever we use the items() method it returns the keys and values in a tuple
# We can easily unpack this in the header of the for loop as follows
for key,value in D.items():
    print(f"fruit is {key} and it costs {value}")
```
