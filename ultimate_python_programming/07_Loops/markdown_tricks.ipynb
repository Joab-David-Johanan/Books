***Some Markdown tricks:***

`Italics and Bold text`

- single pair of asterik signs (* text *) is used for making text `italic` -> *italic text*
- single pair of underscore signs (_ text _) is used for making text `italic` -> _italic text_
<!---->
- double pair of asterik signs (** text **) is used for making text `bold` -> **bold text**
- double pair of underscore signs (__ text __) is used for making text `bold` -> __bold text__
<!---->
- triple pair of asterik signs (*** text ***) is used for making text `italic and bold` -> ***italic and bold text***
- triple pair of underscore signs (___ text ___) is used for making text `italic and bold` -> ___italic and bold text___
